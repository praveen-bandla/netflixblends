```python
import json
import pandas as pd
```

```python
def read_metadata():
    lines = []
    for line in open('movielens_data_genome\\movie_dataset_public_final\\raw\\metadata_updated.json', 'r'):
        lines.append(json.loads(line))
    return lines
```

```python
lines = read_metadata()
```

```python
lines
```

```python
df = pd.DataFrame(columns = ['movie_title', 'imdb_id', 'item_id', 'year'])
```

```python
for line in lines:
    df.loc[len(df.index)] = [line['title'][:-7], line['imdbId'], line['item_id'], line['title'][-6:]]
```

```python
df
```

```python
df.to_csv('titles_movielens.csv')
```
