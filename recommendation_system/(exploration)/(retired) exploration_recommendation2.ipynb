```python
import pandas as pd
import numpy as np
```

```python
#file with all label values for each movie
df = pd.read_csv('movielens_data_genome/movie_dataset_public_final/scores/glmer.csv')
#currently only has 9734 movies
titles = df['item_id'].unique()
```

```python
df
```

```python
#generating user watch history based on random for now. Need to add actual history later
sample_user = np.random.choice(titles, 20)
```

```python
sample_user
```

```python
# creating a matrix where (in order of the titles in sample_user) each row corresponds to the tag values of the given movie title

sample_user_matrix = []

for movie in sample_user:
    #skip if for some reason the movie does not contain all labels
    if len(df[df.item_id == movie]) != 1084:
        print(f'error {movie}')
    
    #if the movie contains all label values, obtain the tag vector from the main df
    else:
        matrix = list(df[df.item_id == movie].score.values)
        sample_user_matrix.append(matrix)

#turning the list of lists into a matrix as a 2D array
mat = np.array(sample_user_matrix)
```

```python
mat
```

```python
matsq = mat**2
agg = np.average(matsq, axis = 0)
aggroot = agg**0.5
```

```python
master_mat = []

for movie in titles:
    if len(df[df.item_id == movie]) != 1084:
        print(f'error {movie}')

    else:
       matrix = list(df[df.item_id == movie].score.values)
       master_mat.append(matrix)
master_mat = np.array(master_mat)
```

```python
rows, cols = master_mat.shape
dist = []
for r in range(rows):
    d = np.linalg.norm(aggroot-master_mat[r])
    dist.append(d)
```

```python
dist
```

```python
dfresult = pd.DataFrame(titles[0:-1], columns = ['movie_id'])
```

```python
dfresult['score1'] = dist
```

```python
dfresult = dfresult.sort_values(by = ['score1'], ascending = False)
dfresult
```
