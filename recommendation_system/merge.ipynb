This notebook shall handle the merging of the movielens dataset with the Netflix titles dataset

```python
import pandas as pd
```

```python
movielens = pd.read_csv("titles_movielens.csv")
netflix = pd.read_csv("../Netflix-Data/titles.csv")
netflix = netflix[netflix['type'] == 'MOVIE']
```

```python
movielens[movielens['movie_title'] == 'Taxi Driver']
```

```python
netflix.head()
```

```python
df = netflix.merge(movielens, left_on='title', right_on='movie_title',
                    how='inner', suffixes=('_x', ''))
```

```python
cols = ['Unnamed: 0','movie_title', 'imdb_id', 'item_id', 'year']
df = df[cols].drop_duplicates('item_id')
```

```python
df
```

```python
df.to_csv("titles_movielens_netflix.csv", index=False)
```
